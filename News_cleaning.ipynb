```python
import pandas as pd

# Define the path to your Excel file
file_path = '/content/drive/MyDrive/narxoz/cryptopanic_news.xlsx'

# Load the dataset from the provided path, specifying that there are no headers
df = pd.read_excel(file_path, header=None)

# Manually assign the column names
df.columns = [
    'id', 'title', 'sourceId', 'newsDatetime', 'url','s', 'negative', 'positive',
    'important', 'liked', 'disliked', 'lol', 'toxic', 'saved', 'comments'
]

df

# Save the dataframe to a CSV file if needed
#df.to_csv('cryptopanic_news.csv', index=False)
```

```python
df_dropped = df.drop(columns=['s'])
```

```python
start_date = '2024-01-01'
end_date = '2025-01-01'

date_filtered_df = df_dropped[(df_dropped['newsDatetime'] >= start_date) & (df_dropped['newsDatetime'] <= end_date)]

# Apply the filter for at least one occurrence in the 'important' column
filtered_df = date_filtered_df[date_filtered_df['important'] > 0]
```

```python
df = pd.read_csv('/content/drive/MyDrive/narxoz/news__currency.csv')
```

```python
df.columns = ["news id", 'currency id']
df
```

```python
currencies = pd.read_csv('/content/drive/MyDrive/narxoz/currency.csv')
currencies.columns = ['currency id', 'currency name', '1','2','3','4','5','6']
currencies = currencies.drop(columns=['1','2','3','4','5','6'])
currencies
```

```python
filtered_df = filtered_df.merge(df, left_on='id', right_on='news id')

# Merge the result with the currencies dataset to add currency names
filtered_df = filtered_df.merge(currencies, left_on='currency id', right_on='currency id')

# Display the final DataFrame
#print(filtered_df)

# Optionally, display the first few rows to verify
#print(filtered_df.head())

# Save the final DataFrame to a CSV file if needed
#filtered_df.to_csv('cryptopanic_news_with_currencies.csv', index=False)
```

```python
filtered_df
```

```python
final_df = filtered_df[['title', 'newsDatetime', 'currency name']]
```

```python
crypto_names = ['BTC', 'ETH', 'BNB', 'XRP', 'ADA', 'SOL', 'DOGE', 'DOT', 'SHIB', 'LTC']
final_df = final_df[filtered_df['currency name'].isin(crypto_names)]
```

```python
final_df
```

```python
from transformers import pipeline
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed


# Initialize the sentiment analysis pipeline
sentiment_model = pipeline(model="ahmedrachid/FinancialBERT-Sentiment-Analysis")

# Batch size for sentiment analysis
batch_size = 50

# Function to process a batch of titles
def analyze_sentiment_batch(batch):
    results = sentiment_model(batch)
    sentiments = [result['label'] for result in results]
    scores = [result['score'] for result in results]
    return sentiments, scores

# Process sentiment analysis in parallel
sentiments = []
scores = []

with ThreadPoolExecutor() as executor:
    futures = []
    for i in range(0, len(final_df), batch_size):
        batch = final_df['title'][i:i + batch_size].tolist()
        futures.append(executor.submit(analyze_sentiment_batch, batch))

    for future in tqdm(as_completed(futures), total=len(futures)):
        batch_sentiments, batch_scores = future.result()
        sentiments.extend(batch_sentiments)
        scores.extend(batch_scores)

# Add sentiment and score to the DataFrame
final_df['sentiment'] = sentiments
final_df['score'] = scores
```
